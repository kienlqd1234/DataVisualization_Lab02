### Dữ liệu có các cột bị lặp không?

- Quan sát số liệu năm 2018, bạn có thể thấy các cột `Accident Fields_Reference Number`, `Vehicle Fields_Reference Number` và `Casualty Fields_Reference Number` dường như chứa dữ liệu khá giống nhau.

- Ngoài ra, cột  `Number of Vehicles` và `Vehicle Number` trong `data_2018` và `data_2019` cũng trông khá giống nhau.

- Để chắc chắn về các giả thuyết này, nhóm sẽ tiến hành kiểm tra các cột này bằng cách thiết kế hàm `check_similarity()` để kiểm tra 2 cột theo tỉ lệ phần trăm giống nhau.

```python
def check_similarity(df, col_name_1, col_name_2):
    # So sánh trực tiếp
    comparison = df[col_name_1] == df[col_name_2]

    # Tính tỷ lệ phần trăm giống nhau
    similarity_percentage = comparison.mean()*100
    
    print(f'Tỷ lệ phần trăm giống nhau giữa {col_name_1} và {col_name_2}: {similarity_percentage:.2f}%')

check_similarity(df1, 'Accident Fields_Reference Number', 'Vehicle Fields_Reference Number')
check_similarity(df1, 'Accident Fields_Reference Number', 'Casualty Fields_Reference Number')
check_similarity(df1, 'Number of Vehicles', 'Vehicle Number')
check_similarity(df2, 'Number of Vehicles', 'Vehicle Number')
```

- Theo kết quả trên, cột `Accident Fields_Reference Number`, `Vehicle Fields_Reference Number` và `Casualty Fields_Reference Number` chứa cùng một dữ liệu. Chúng ta tiến hành loại bỏ 2 trong số chúng.
- Có thể hơi nhiễu khi tiế hành phân tích 2 cột có độ tương tự hơn 65% nên nhóm cũng quyết định xóa cột `Vehicle Number`.

### Sự phân bố giá trị của mỗi cột

#### Với mỗi cột có kiểu dữ liệu dạng numeric, các giá trị được phân bố như thế nào?

- Với các cột có kiểu dữ liệu số, ta tính:

    - Tỉ lệ % (từ 0 đến 100) các giá trị thiếu
    - Giá trị min
    - Giá trị lower quartile (phân vị 25)
    - Giá trị median (phân vị 50)
    - Giá trị upper quartile (phân vị 75)
    - Giá trị max

- Nhóm sẽ lưu kết quả vào DataFrame `num_col_info_df`, trong đó:
    - Tên các cột là tên các cột số trong `raw_df`
    - Tên các hàng: “missing_ratio”, “min”, “low_quartile”, “median”, “upper_quartile”, “max”

- Để dễ quan sát, nhóm làm tròn tất cả các giá trị đến 1 chữ số thập phân bằng phương thức `.round(1)`.

```python
num_col_info_df = None

# Define missing ratio
def missing_ratio(series):
    percentage = series.isna().sum() / len(series)
    return percentage.round(1)

# Define lower quartile
def lower_quartile(series):
    return series.quantile(0.25).round(1)

# Define upper quartile
def upper_quartile(series):
    return series.quantile(0.75).round(1)

#Define median
def median(series):
    return series.quantile(0.75).round(1)

num_col_info_df = df.select_dtypes(include=['int64', 'float64'])

# Ta sẽ không khảo sát phân bố của các cột tọa độ
num_col_info_df = num_col_info_df.drop(columns=["Grid Ref: Easting", "Grid Ref: Northing"])

num_col_info_df = num_col_info_df.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])

num_col_info_df
```

##### Nhận xét
- Các chỉ số `min` và `max` cho thấy số lượng phương tiện tham gia gặp tai nạn trải rộng từ 1 đến 7 và độ tuổi của thương vong trải rộng từ 1 đến 95.
- Từ giá trị tứ phân vị ta có thể thấy với cột `Number of Vehicles`, 75% giá trị là 1 hoặc 2 và chỉ có 25% giá trị phân bố từ 3 đến 7. Tương tự với cột `Age of Casualty` thì 75% giá trị độ tuổi là 1 đến 49 và chỉ có 25% giá trị phân bố từ 50 đến 95

#### Với mỗi cột có kiểu dữ liệu dạng categorical, các giá trị được phân bố như thế nào?

- Với các cột có kiểu dữ liệu dạng `categorical`, ta cần tính:

    - Tỉ lệ % (từ 0 đến 100) các giá trị thiếu
    - Số lượng các giá trị khác nhau (không xét giá trị thiếu)
    - Tỉ lệ % (từ 0 đến 100) của mỗi giá trị được sort theo tỉ lệ % giảm dần (không xét giá trị thiếu)

- Chúng ra sẽ lưu kết quả vào DataFrame `cat_col_info_df`, trong đó:
    - Tên các cột là tên các cột không phải số trong `raw_df`
    - Tên các dòng là: “missing_ratio”, “num_values”, “value_ratios”

- Để tiện quan sát, chúng ta làm tròn tất cả các giá trị đến 1 chữ số thập phân bằng phương thức `.round(1)`.

```python
pd.set_option('display.max_colwidth', 100) # For clearly
pd.set_option('display.max_columns', None) # For clearly
```

```python
cat_col_info_df = df.select_dtypes(include='object')

# Define num values
def num_values(series):
    return len(series.dropna().unique())

# Define value ratios
def value_ratios(series):
    value_counts = series.dropna().value_counts()
    percentages = (value_counts / len(series))*100
    percentages = percentages.round(1)
    sorted_percentages = percentages.sort_values(ascending=False)
    return sorted_percentages.to_dict()
cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df
```
